```python
#Import necessary libraries
import numpy as np
import pandas as pd
```

```python
#Load the dataset into a Pandas DataFrame
df = pd.read_csv('orders.csv')
```

```python
#Display the first few rows
df.head()
```

```python
df.info()
```

```python
#columns in df
df.columns
```

```python
#Rename columns for better readability and consistency
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df.columns
```

```python
#Converting the 'order_date' column to datetime format
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
#data after convert
df.info()
```

```python
#Checking for missing values in the dataset
df.isnull().sum()
```

```python
df.ship_mode.value_counts()
```

```python
#Replace 'Not Available' and 'unknown' in the 'ship_mode' column with None.
df['ship_mode'].replace(['Not Available','unknown'], np.nan, inplace=True)
```

```python
df.ship_mode.value_counts()
```

```python
#Verify that the replacement worked
df.isnull().sum()
```

```python
#create new columns

# Creating a new column 'discount' to calculate the discount amount
df['discount'] = df['list_price']*df['discount_percent']/100

#Create 'sales_price' by subtracting the discount from the list price
df['sales_price'] = df['list_price'] - df['discount']

# Calculate 'profit' as the difference between sales price and cost price
df['profit'] = df['sales_price'] - df['cost_price']

df.head(10)
```

```python
#dropping irrelevant columns
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
```

```python
#Checking for duplicate values
df.duplicated().sum()
```

```python
#storing the cleaned data in new csv file
cleaned_data = df.to_csv('cleaned_data.csv',index=False)
```
